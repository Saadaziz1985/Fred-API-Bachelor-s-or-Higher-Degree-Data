```python
#Import all packages & libraries incluing Fred API
import pandas as pd
from fredapi import Fred
import matplotlib.pyplot as plt
fred = Fred(api_key='Enter_Key_Here')
```

```python
#Search the Fred, initially with two filters, then add two more filters 
#and lastly to make sure title contains the required result
degree_states_df = fred.search("Bachelor's Degree or Higher for", filter=('frequency','Annual'))
degree_states_df = degree_states_df.query('seasonal_adjustment_short == "NSA" and units == "Percent"')
degree_states_df = degree_states_df.loc[degree_states_df['title'].str.contains("Bachelor's Degree or Higher for")]
```

```python
degree_states_df
```

```python
#Loop is used to extract data and timer used to delay data extraction to avoid blocked in Fred website
# Appemd and concat used to combine all results in all_states_results column
import time
all_state_results = []

for s_id in degree_states_df.index:
    state_results = fred.get_series(s_id)
    state_results = state_results.to_frame(name=s_id)
    all_state_results.append(state_results)
    time.sleep(0.1) #It will take long time to avoid blocked in the Fred website
degree_results_states = pd.concat(all_state_results, axis=1) #Concatenate all result as columns
```

```python
#Extract all data based on states codes
degree_results_states
```

```python
# Below is used to get the names of the states
degree_states = degree_results_states.copy()  #.drop('UNRATE', axis=1)
degree_states = degree_states.dropna()
state_id = degree_states_df['title'].str.replace("Bachelor's Degree or Higher for",'').to_dict()
degree_states.columns = [state_id[c] for c in degree_states.columns]
```

```python
degree_states
```

```python
#Export file to csv format for all States.
degree_states.to_csv("Bachelor's Degree or Higher for USA States.csv")
```

```python
degree_states.plot(figsize=[12,8])
plt.xlabel("Year")
plt.ylabel("Percentage of Bachelor's Degree or Higher for USA States")
plt.title("Year & Percentage of Bachelor's Degree or Higher for USA States")
plt.show()
```
